## 1. Matrix multiplication and noncommutativity

Compute `A @ B` and `B @ A`. Are they the same?

```python
A = np.array([[1, 2],
              [3, 4]], dtype=float)

B = np.array([[1, 1],
              [0, 1]], dtype=float)

print('AB =')
print(A @ B)
print('\nBA =')
print(B @ A)
print('\nAB equals BA?', np.allclose(A @ B, B @ A))
```

## 2. Inverse, determinant, and rank

A square matrix is invertible exactly when it has full rank. For a \(2\times2\) matrix, this is equivalent to nonzero determinant.

```python
A = np.array([[4, 7],
              [2, 6]], dtype=float)

print('det(A) =', np.linalg.det(A))
print('rank(A) =', np.linalg.matrix_rank(A))
print('A inverse =')
print(np.linalg.inv(A))
print('A @ A_inverse =')
print(A @ np.linalg.inv(A))
```

## 3. Transpose and Gram matrix

The Gram matrix \(A^T A\) records all pairwise dot products among the columns of \(A\).

```python
X = np.array([[1, 2],
              [3, 4],
              [5, 6]], dtype=float)

G = X.T @ X
print('X^T X =')
print(G)
print('Is symmetric?', np.allclose(G, G.T))
```

## 4. LU factorization using SymPy

SymPy can compute exact LU factorizations.

```python
A = sp.Matrix([[2, 1, 1],
               [4, -6, 0],
               [-2, 7, 2]])

L, U, perm = A.LUdecomposition()
print('L =')
display(L)
print('U =')
display(U)
print('Permutation data =', perm)
print('Check L*U =')
display(L * U)
```

## 5. Computational cost experiment

Compare `(A @ B) @ x` with `A @ (B @ x)`. They are mathematically the same, but can have very different computational cost.

```python
import time

m, n, p = 600, 400, 300
A = np.random.randn(m, n)
B = np.random.randn(n, p)
x = np.random.randn(p)

t0 = time.time()
y1 = (A @ B) @ x
t1 = time.time()

t2 = time.time()
y2 = A @ (B @ x)
t3 = time.time()

print('Time for (A @ B) @ x:', t1 - t0)
print('Time for A @ (B @ x):', t3 - t2)
print('Same answer?', np.allclose(y1, y2))
```
