## Areas, Determinants, and the Largest Inscribed Ellipse

**Purpose.** This independent-study notebook explains why determinants measure area scaling and uses this idea to solve a geometric problem about triangles and ellipses.

The central theorem is:

> If $L:\mathbb R^2\to\mathbb R^2$ is represented by a matrix $A$, then for any region $C$ in the plane,
>
> $$
> \operatorname{Area}(L(C))=|\det(A)|\operatorname{Area}(C).
> $$

## 1. Why determinants scale area

Suppose a parallelogram is spanned by vectors $u,v$. Its area is

$$
|\det[u\ v]|.
$$

If $L(x)=Ax$, then the image parallelogram is spanned by $Au,Av$, so its area is

$$
|\det[Au\ Av]|
=
|\det(A[u\ v])|
=
|\det(A)|\,|\det[u\ v]|.
$$

Thus the area is multiplied by $|\det(A)|$.

This extends from parallelograms to general regions by approximation or by the Jacobian formula.

```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
import math

sqrt3 = sp.sqrt(3)

# Vertices of triangle C
C0 = sp.Matrix([0,0])
C1 = sp.Matrix([2,0])
C2 = sp.Matrix([1,sqrt3])

# Vertices of triangle D
D0 = sp.Matrix([0,0])
D1 = sp.Matrix([3,0])
D2 = sp.Matrix([3,4])
```

## 2. The two triangles

Triangle $C$ is equilateral with vertices

$$
(0,0),\qquad (2,0),\qquad (1,\sqrt3).
$$

Triangle $D$ is the $3$-$4$-$5$ right triangle with vertices

$$
(0,0),\qquad (3,0),\qquad (3,4).
$$

We want a linear map $L$ such that

$$
L(C)=D.
$$

```python
C_np = np.array([[0,0],[2,0],[1,float(sp.sqrt(3))],[0,0]])
D_np = np.array([[0,0],[3,0],[3,4],[0,0]])

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.plot(C_np[:,0], C_np[:,1], marker="o")
plt.title("Triangle C")
plt.axis("equal")
plt.grid(True)

plt.subplot(1,2,2)
plt.plot(D_np[:,0], D_np[:,1], marker="o")
plt.title("Triangle D")
plt.axis("equal")
plt.grid(True)

plt.show()
```

## 3. Task 1: Find the matrix $A$

A linear map sends $0$ to $0$. We choose the map so that the two sides of $C$ meeting at the origin map to the two sides of $D$ meeting at the origin:

$$
A\begin{bmatrix}2\\0\end{bmatrix}
=
\begin{bmatrix}3\\0\end{bmatrix},
\qquad
A\begin{bmatrix}1\\\sqrt3\end{bmatrix}
=
\begin{bmatrix}3\\4\end{bmatrix}.
$$

In matrix form:

$$
A
\begin{bmatrix}
2&1\\
0&\sqrt3
\end{bmatrix}
=
\begin{bmatrix}
3&3\\
0&4
\end{bmatrix}.
$$

So

$$
A=
\begin{bmatrix}
3&3\\
0&4
\end{bmatrix}
\begin{bmatrix}
2&1\\
0&\sqrt3
\end{bmatrix}^{-1}.
$$

```python
C_basis = sp.Matrix.hstack(C1, C2)
D_basis = sp.Matrix.hstack(D1, D2)

A = sp.simplify(D_basis * C_basis.inv())
A_inv = sp.simplify(A.inv())

print("A =")
display(A)

print("det(A) =")
display(sp.factor(A.det()))

print("A^{-1} =")
display(A_inv)

print("A*C1 =")
display(A*C1)

print("A*C2 =")
display(A*C2)
```

Therefore

$$
A=
\begin{bmatrix}
\frac32 & \frac{\sqrt3}{2}\\[4pt]
0&\frac{4\sqrt3}{3}
\end{bmatrix},
\qquad
\det(A)=2\sqrt3.
$$

## 4. Task 2: Largest circle inscribed in $C$

For an equilateral triangle, the largest inscribed circle is the incircle.

Triangle $C$ has side length $2$. The altitude is $\sqrt3$, and the inradius is one-third of the altitude:

$$
r_C=\frac{\sqrt3}{3}.
$$

The center is the centroid/incenter

$$
(a,b)=\left(1,\frac{\sqrt3}{3}\right).
$$

Thus the area of the largest circle is

$$
\pi r_C^2
=
\pi\left(\frac{\sqrt3}{3}\right)^2
=
\frac{\pi}{3}.
$$

```python
rC = sqrt3/3
area_circle_C = sp.pi * rC**2

print("Inradius of C =", rC)
print("Area of incircle =", sp.simplify(area_circle_C))
```

```python
theta = np.linspace(0, 2*np.pi, 300)
center_C = np.array([1, float(sp.sqrt(3)/3)])
radius_C = float(sp.sqrt(3)/3)
circle = center_C + radius_C*np.column_stack([np.cos(theta), np.sin(theta)])

plt.figure(figsize=(6,6))
plt.plot(C_np[:,0], C_np[:,1], marker="o")
plt.plot(circle[:,0], circle[:,1])
plt.scatter([center_C[0]], [center_C[1]])
plt.text(center_C[0]+0.05, center_C[1]+0.05, "center")
plt.title("Largest Circle Inscribed in C")
plt.axis("equal")
plt.grid(True)
plt.show()
```

## 5. Task 3: Area of the largest ellipse inscribed in $D$

A nonsingular linear map sends ellipses to ellipses. Since $D=L(C)$, the image of the incircle of $C$ is an ellipse inscribed in $D$.

Its area is scaled by

$$
|\det(A)|=2\sqrt3.
$$

Therefore the area of the ellipse is

$$
\operatorname{Area}
=
|\det(A)|\cdot \frac{\pi}{3}
=
2\sqrt3\cdot \frac{\pi}{3}
=
\frac{2\pi\sqrt3}{3}
=
\frac{2\pi}{\sqrt3}.
$$

```python
area_ellipse_D = sp.simplify(abs(A.det()) * area_circle_C)
print("Area of largest ellipse in D =")
display(area_ellipse_D)
```

This is also consistent with the general theorem that every triangle has a unique maximum-area inscribed ellipse, called the **Steiner inellipse**, and for an equilateral triangle this ellipse is the incircle.

## 6. Task 4: Equation of the largest ellipse in $D$

The incircle in $C$ has equation

$$
(x-1)^2+\left(y-\frac{\sqrt3}{3}\right)^2
=
\left(\frac{\sqrt3}{3}\right)^2.
$$

The image point in $D$ is

$$
\begin{bmatrix}u\\v\end{bmatrix}
=
A
\begin{bmatrix}x\\y\end{bmatrix}.
$$

Therefore

$$
\begin{bmatrix}x\\y\end{bmatrix}
=
A^{-1}
\begin{bmatrix}u\\v\end{bmatrix}.
$$

Since

$$
A^{-1}
=
\begin{bmatrix}
\frac23&-\frac14\\[4pt]
0&\frac{\sqrt3}{4}
\end{bmatrix},
$$

we have

$$
x=\frac23u-\frac14v,
\qquad
y=\frac{\sqrt3}{4}v.
$$

Substitute into the circle equation:

$$
\left(\frac23u-\frac14v-1\right)^2
+
\left(\frac{\sqrt3}{4}v-\frac{\sqrt3}{3}\right)^2
=
\frac13.
$$

```python
u, v = sp.symbols("u v")
x_expr = sp.Rational(2,3)*u - sp.Rational(1,4)*v
y_expr = sp.sqrt(3)/4*v

ellipse_eq = sp.expand((x_expr - 1)**2 + (y_expr - sp.sqrt(3)/3)**2 - sp.Rational(1,3))
ellipse_eq_simplified = sp.factor(ellipse_eq)

print("Ellipse equation in implicit form equals 0:")
display(ellipse_eq_simplified)

print("Cleared denominator form:")
display(sp.expand(36*ellipse_eq_simplified))
```

Equivalently,

$$
\frac{16u^2-12uv-48u+9v^2+36}{36}=0,
$$

or

$$
16u^2-12uv+9v^2-48u+36=0.
$$

This is the equation of the largest ellipse inscribed in triangle $D$.

```python
theta = np.linspace(0, 2*np.pi, 500)
circle_pts = np.vstack([
    1 + radius_C*np.cos(theta),
    float(sp.sqrt(3)/3) + radius_C*np.sin(theta)
])
A_np = np.array(A.evalf(), dtype=float)
ellipse_pts = A_np @ circle_pts

plt.figure(figsize=(6,6))
plt.plot(D_np[:,0], D_np[:,1], marker="o", label="Triangle D")
plt.plot(ellipse_pts[0,:], ellipse_pts[1,:], label="Image ellipse")
plt.title("Largest Ellipse Inscribed in D")
plt.axis("equal")
plt.grid(True)
plt.legend()
plt.show()
```

## 7. Additional verification: area by singular values

A linear map sends a circle of radius $r$ to an ellipse. The semiaxis lengths are

$$
r\sigma_1,\qquad r\sigma_2,
$$

where $\sigma_1,\sigma_2$ are the singular values of $A$.

The area is

$$
\pi(r\sigma_1)(r\sigma_2)
=
\pi r^2(\sigma_1\sigma_2)
=
\pi r^2|\det(A)|.
$$

This agrees with the determinant area scaling formula.

```python
A_float = np.array(A.evalf(), dtype=float)
singular_values = np.linalg.svd(A_float, compute_uv=False)

print("Singular values of A:", singular_values)
print("Product of singular values:", singular_values.prod())
print("|det(A)|:", abs(float(A.det())))

semi_axes = radius_C * singular_values
print("Ellipse semi-axis lengths:", semi_axes)
print("Area from semi-axes:", math.pi * semi_axes[0] * semi_axes[1])
print("Area from determinant:", float(area_ellipse_D))
```

## 8. Independent-study checkpoints

### Checkpoint A
Why does a linear map send a triangle to a triangle?

**Answer.** A line segment from $u$ to $v$ is the set
$$
tu+(1-t)v,\qquad 0\le t\le 1.
$$
A linear map sends this to
$$
tL(u)+(1-t)L(v),
$$
which is the line segment from $L(u)$ to $L(v)$. Therefore the edges of a triangle are sent to line segments.

### Checkpoint B
Why do we use $|\det(A)|$ instead of $\det(A)$?

**Answer.** Area is nonnegative. A negative determinant reverses orientation, but it still scales area by the absolute value.

### Checkpoint C
Why does the incircle become an ellipse?

**Answer.** The circle equation becomes a quadratic equation after substituting $x,y$ as linear functions of $u,v$. A nonsingular linear map sends circles to ellipses.
